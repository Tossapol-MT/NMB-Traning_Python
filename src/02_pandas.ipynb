## Concat

### Reset Index

```python
df_reset_index = df_total.reset_index(drop=True)
df_reset_index
```

### Drop duplicate

```python
boolean = df_reset_index['emp_name'].duplicated().any()
print(boolean)
print(len(df_reset_index))

df_dup = df_reset_index.drop_duplicates(['emp_name'])
print(len(df_dup))

boolean = df_dup['emp_name'].duplicated().any()
print(boolean)
df_dup
```

### Drop Null

```python
print(df_dup.isnull().values.any())
print(len(df_dup))

df_null = df_dup.dropna()
print(len(df_null))

df_null
```

#### Calculate df

```python
import datetime

year = datetime.date.today().year
print(year)

df_age = df_null.assign(age=year-df_null["birthday year"])
df_age
```

```python
df_age[["name","sername"]] = df_age["emp_name"].str.split(expand=True)
df_age
```

### Custom Col

```python
df_col = df_age[["name","sername","age","country"]]
df_col
```

#### Edit columns name

```python
df_edit_name = df_col.rename(columns={'name':"first_name",'sername':"last_name"})
df_edit_name
```

```python
def csv_sort_country(df_split_contry):
    df_unique = df_split_contry["country"].unique()
    for i in df_unique:
        new_file = df_split_contry[df_split_contry["country"] == i].to_csv(f"export\df_sorted_{i}.csv")
    return new_file
```

```python
import xlwings as xw

def export_sorted_excel_one_file(df):
    app = xw.App(visible=False)
    template = xw.Book()
    sheet = template.sheets('Sheet1')
    country_list = df["country"].unique()
    
    for i in range(len(country_list)):
        template.sheets.add(country_list[i])
        sheet = template.sheets(country_list[i])
        df_result = df[df["country"] == country_list[i]]
        # sheet["A1"].value = df_result
        sheet.range("A1").options(pd.DataFrame, index = False).value = df_result #Delete index
        
    for sheet in template.sheets:
        if 'Sheet1' in sheet.name:
            sheet.delete()
            
    template.save(r"export\data_with_xlwings.xlsx")
    template.close()
    app.quit()
    
export_sorted_excel_one_file(df_lookup)
    
```
